```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.signal import savgol_filter

# Fungsi untuk membaca file .log (dipisahkan oleh tab atau spasi)
def read_log_file(file_path):
    return pd.read_csv(file_path, sep=r'\s+', engine='python')

# Load data dari file log
train_log = read_log_file("results/log/train.log")
val_log = read_log_file("results/log/val.log")
#batch_log = read_log_file("results/log/train_batch.log")

# Fungsi smoothing menggunakan Savitzky-Golay filter
def smooth_data(y, window=11, poly=2):
    return savgol_filter(y, window_length=min(window, len(y)), polyorder=poly)  # Hindari error jika data pendek

# Pastikan ukuran plot seragam
plt.rcParams.update({'figure.figsize': (10, 5)})

# ====== PLOT TRAINING vs VALIDATION LOSS ======
plt.figure()
sns.lineplot(x=train_log['epoch'], y=smooth_data(train_log['loss']), label="Train Loss", marker='o', linestyle='-')
sns.lineplot(x=val_log['epoch'], y=smooth_data(val_log['loss']), label="Validation Loss", marker='s', linestyle='--')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ====== PLOT TRAINING vs VALIDATION ACCURACY ======
plt.figure()
sns.lineplot(x=train_log['epoch'], y=smooth_data(train_log['acc']), label="Train Accuracy", marker='o', linestyle='-')
sns.lineplot(x=val_log['epoch'], y=smooth_data(val_log['acc']), label="Validation Accuracy", marker='s', linestyle='--')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ====== PLOT BATCH-WISE LOSS PER EPOCH ======
#plt.figure(figsize=(12,6))
#sns.lineplot(
#    x=batch_log['iter'], 
#    y=smooth_data(batch_log['loss'], window=21),  # Smoothing lebih besar untuk batch log
#    hue=batch_log['epoch'], 
#    marker="o", 
#    palette="tab10",
#    alpha=0.7
#)
#plt.xlabel("Iteration")
#plt.ylabel("Loss")
#plt.title("Batch-wise Loss per Epoch")
#plt.legend(title="Epoch", bbox_to_anchor=(1, 1))
#plt.grid(alpha=0.3)
#plt.show()
```
